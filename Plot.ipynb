```python
import torch
import numpy
import numpy as np

n_m = 10
n_n = 10
nrep = 20
```

```python
import os
os.getcwd()
```

## Experiemnt 1

### Participation Error

```python
e = torch.load('e_1_p_error_30_20.pt',map_location=torch.device('cpu'))
e = np.log(e)
#e = np.mean(e.numpy(),2)
mat = np.empty([n_m * n_n*nrep, 2])
#error = np.std(e_10.numpy(),2)
i = 0
for ct_m in range(n_m):
    for ct_n in range(n_n):
        for rep in range(nrep):
            m = (ct_m+1)*20
            n = (ct_n+1)*20
            mat[i,0] = m * n
            mat[i,0] = np.log(mat[i,0])
            mat[i,1] = e[ct_m,ct_n,rep]
            i += 1
            
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 15

x = mat[:,0]
y = mat[:,1]

x_unique  = np.unique(x)
y_means = np.array([np.mean(y[x==u]) for u in x_unique])
y_stds = np.array([np.std(y[x==u]) for u in x_unique])

x = x_unique
y = y_means
error = y_stds

sorted_indices = np.argsort(x)
x = x[sorted_indices]

y = y[sorted_indices]

error = error[sorted_indices]
plt.plot(x, y, marker = 'o', label = 'd_int = 20')
#plt.scatter(x,y)
plt.fill_between(x, y - error, y + error, alpha=0.2)
plt.plot(x, y+error, linestyle = 'dashed', c = 'black', alpha = 0.5)
plt.plot(x, y-error, linestyle = 'dashed', c = 'black', alpha = 0.5)

#################################################

e = torch.load('e_1_p_error_30_10.pt', map_location=torch.device('cpu'))
e = np.log(e)
#e = np.mean(e.numpy(),2)
mat = np.empty([n_m * n_n*nrep, 2])
#error = np.std(e_10.numpy(),2)
i = 0
for ct_m in range(n_m):
    for ct_n in range(n_n):
        for rep in range(nrep):
            m = (ct_m+1)*20
            n = (ct_n+1)*20
            mat[i,0] = m * n
            mat[i,0] = np.log(mat[i,0])
            mat[i,1] = e[ct_m,ct_n,rep]
            i += 1
            

x = mat[:,0]
y = mat[:,1]

x_unique  = np.unique(x)
y_means = np.array([np.mean(y[x==u]) for u in x_unique])
y_stds = np.array([np.std(y[x==u]) for u in x_unique])

x = x_unique
y = y_means
error = y_stds

sorted_indices = np.argsort(x)
x = x[sorted_indices]

y = y[sorted_indices]

error = error[sorted_indices]
plt.plot(x, y, marker = 'o', label = 'd_int = 10')
#plt.scatter(x,y)
plt.fill_between(x, y - error, y + error, alpha=0.2)
plt.plot(x, y+error, linestyle = 'dashed', c = 'black', alpha = 0.5)
plt.plot(x, y-error, linestyle = 'dashed', c = 'black', alpha = 0.5)
plt.xlabel("Log (mn)")
plt.ylabel("Log (MSE)")

plt.legend(loc='upper center', bbox_to_anchor=(0.5,1.2),ncol=2)
#plt.errorbar(x, y, yerr=error, fmt='o', capsize = 3)
plt.savefig('P_1.pdf', bbox_inches='tight')
```
